```python
import numpy as np  
import matplotlib.pyplot as plt  
import math
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = eval(formula)
    plt.plot(x, y)  
    plt.show()
```

```python
delaT = 1  * 3.154* (10**7)
H = 997 * 4187 * 300
lamb = 1.25
f= 3.7
T = [0]
for i in range(400):
    Result = T[i] + (delaT / H)*(f - lamb*T[i])
    T.append(Result)
plt.plot(range(401),T)
plt.xlabel('year',fontsize=16, fontweight='bold')
plt.ylabel('temperature',fontsize=16, fontweight='bold')
plt.plot(range(401), [3]*401, c='black',linewidth=3.0)
```

```python
f = lambda x: 0.037*x if x < 100 else 3.7
H = 997 * 4187 * 300
T = [0]
delaT = 1  * 3.154* (10**7)
plt.plot(range(401), [3]*401, c='black',linewidth=3.0)
for i in range(400):
    Result = T[i] + (delaT / H)*(f(i) - lamb*T[i])
    T.append(Result)
plt.plot(range(401),T ,label="graph1")
plt.xlabel('year',fontsize=16, fontweight='bold')
plt.ylabel('temperature',fontsize=16, fontweight='bold')
result = dict(zip(range(401),T))
f1 = lambda x: 0.037*x if x <= 98 else 0
T1 = [0]
for i in range(400):
    Result = T1[i] + (delaT / H)*(f1(i) - lamb*T1[i])
    T1.append(Result)
plt.plot(range(401),T1,label="graph2")
Tau_Diff = 10
f2 = lambda x: 0.037*x if x <=98 else f2(98)*math.exp((98-x)/Tau_Diff)
T2 = [0]
plt.plot(range(401), [3]*401, c='black',linewidth=3.0)
for i in range(400):
    Result = T2[i] + (delaT / H)*(f2(i) - lamb*T2[i])
    T2.append(Result)
plt.plot(range(401),T2,label="graph3")
plt.legend()
```
